```python
import pandas as pd
```

```python
# Loading standardized dataset
data = pd.read_csv("../04_Clean_Data/master_dataset_standardized.csv")
```

```python
data["Sale_Date"] = pd.to_datetime(data["Sale_Date"], format="mixed", dayfirst=True)
data["Procurement_Date"] = pd.to_datetime(data["Procurement_Date"], format="mixed", dayfirst=True)
```

```python
# Convert to standard YYYY-MM-DD format
data["Sale_Date"] = data["Sale_Date"].dt.strftime("%Y-%m-%d")
data["Procurement_Date"] = data["Procurement_Date"].dt.strftime("%Y-%m-%d")
```

```python
data.to_csv("../04_Clean_Data/master_dataset_dates_cleaned.csv", index=False)
```
