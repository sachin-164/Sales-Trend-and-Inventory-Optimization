```python
import pandas as pd
```

```python
sales = pd.read_csv("../01_Original_Files/honeyrich_export_sales.csv")
warehouse = pd.read_csv("../01_Original_Files/honeyrich_supply_logs.csv")
```

Checking relational integrity...


```python
merged = pd.merge(
    sales,
    warehouse,
    on="Batch_ID",
    how="left",
    indicator=True
)

orphan_records = merged[merged["_merge"] == "left_only"]

print("Orphan records:", len(orphan_records))

if len(orphan_records) > 0:
    orphan_records.to_csv("../04_Clean_Data/orphan_records.csv", index=False)

print("Relational integrity verification completed.")
```
