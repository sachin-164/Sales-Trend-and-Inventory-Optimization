```python
import pandas as pd
```

```python
# Loading cleaned dataset
data = pd.read_csv("../04_Clean_Data/master_dataset_dates_cleaned.csv")
```

```python
# Checking weight units
# Ensure all quantities are numeric
data["Quantity_Sold_KG"] = pd.to_numeric(data["Quantity_Sold_KG"], errors="coerce")
```

```python
# If units column exists, convert to KG
if "Unit" in data.columns:
    data["Unit"] = data["Unit"].str.lower().str.strip()

    data.loc[data["Unit"] == "g", "Quantity"] = data["Quantity"] / 1000
    data.loc[data["Unit"] == "g", "Unit"] = "kg"
```

```python
data.to_csv("../04_Clean_Data/master_dataset_final_cleaned.csv", index=False)
```
