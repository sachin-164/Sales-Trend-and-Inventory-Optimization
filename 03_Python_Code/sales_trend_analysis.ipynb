```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv("../04_Clean_Data/master_dataset.csv")
```

```python
# check columns
print(df.columns)
```

```python
# group sales by destination
sales_region = df.groupby("Destination")["Revenue"].sum().sort_values(ascending=False)
print(sales_region)
```

```python
# plot
sales_region.plot(kind="bar", figsize=(8,5))
plt.title("Regional Sales Distribution")
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.show()
```
