```python
import pandas as pd
```

```python
# Loading datasets with encoding verification

sales = pd.read_csv("../01_Original_Files/honeyrich_export_sales.csv", encoding="utf-8")
supply = pd.read_csv("../01_Original_Files/honeyrich_supply_logs.csv", encoding="utf-8")

print("Datasets loaded successfully")
```

```python
# Merging datasets using Batch_ID.

master = pd.merge(
    sales,
    supply,
    on="Batch_ID",
    how="inner"
)
```

```python
# Creating cost and profit metrics

master["Total_Cost"] = master["Quantity_Sold_KG"] * master["Cost_Per_KG"]

master["Profit"] = master["Revenue"] - master["Total_Cost"]
```

```python
# Saving structured master dataset

master.to_csv("../04_Clean_Data/master_dataset.csv", index=False)

print("Master working file created successfully.")
```
