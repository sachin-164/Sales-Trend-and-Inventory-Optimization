```python
import pandas as pd
```

```python
# Loading master dataset
data = pd.read_csv("../04_Clean_Data/master_dataset.csv")
```

```python
# Standardizing product names
data["Product_Name"] = data["Product_Name"].str.strip().str.lower()

# Example corrections
name_corrections = {
    "wild honey": "Wild Honey",
    "forest honey": "Forest Honey",
    "organic honey": "Organic Honey"
}

data["Product_Name"] = data["Product_Name"].replace(name_corrections)
```

```python
data.to_csv("../04_Clean_Data/master_dataset_standardized.csv", index=False)
```
