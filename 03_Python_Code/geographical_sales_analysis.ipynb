```python
# Loading sales dataset
sales = pd.read_csv("../01_Original_Files/honeyrich_export_sales.csv")
```

```python
import pandas as pd

region_sales = (
    sales.groupby("Destination")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

print(region_sales)

region_sales.to_csv("../04_Clean_Data/regional_sales_summary.csv")
```
